```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
retention_ratio = pd.read_csv("D:/김동영/94_sqldata/앱사용자리텐션테이블.csv")
```

```python
retention_ratio
```

```python
# 결측치를 0으로 채움
retention_ratio_filled = retention_ratio.fillna(0)

# cohort_gender를 인덱스로 설정
retention_ratio_filled.set_index('cohort_gender', inplace=True)

# heatmap 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(retention_ratio_filled.iloc[:, 1:], annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Retention Ratio Heatmap")
plt.xlabel("Days")
plt.ylabel("Cohort (Gender)")
plt.show()

```

```python
# Separate the data by gender
retention_ratio_female = retention_ratio_filled[retention_ratio_filled.index.str.contains("\(F\)")]
retention_ratio_male = retention_ratio_filled[retention_ratio_filled.index.str.contains("\(M\)")]


# Plot heatmap for Female
plt.figure(figsize=(12, 8))
sns.heatmap(retention_ratio_female.iloc[:, 1:], annot=True, fmt=".1f", cmap="Reds")
plt.title("Retention Ratio Heatmap (Female)")
plt.xlabel("Days")
plt.ylabel("Cohort (Female)")
plt.show()

# Plot heatmap for Male
plt.figure(figsize=(12, 8))
sns.heatmap(retention_ratio_male.iloc[:, 1:], annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Retention Ratio Heatmap (Male)")
plt.xlabel("Days")
plt.ylabel("Cohort (Male)")
plt.show()
```
