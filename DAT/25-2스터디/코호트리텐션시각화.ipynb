```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
retention_ratio = pd.read_csv("D:/김동영/94_sqldata/앱사용자리텐션테이블.csv")
```

```python
retention_ratio
```

```python
# 결측치를 0으로 채움
retention_ratio_filled = retention_ratio.fillna(0)

# cohort_gender를 인덱스로 설정
retention_ratio_filled.set_index('cohort_gender', inplace=True)

# heatmap 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(retention_ratio_filled.iloc[:, 1:], annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Retention Ratio Heatmap")
plt.xlabel("Days")
plt.ylabel("Cohort (Gender)")
plt.show()

```
