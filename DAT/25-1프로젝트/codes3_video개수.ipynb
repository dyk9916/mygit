```python
# ë¡œì»¬ ê²½ë¡œ ì„¤ì • (ë³¸ì¸ í™˜ê²½ì— ë§ê²Œ ìˆ˜ì •)
base_path = "D:/ê¹€ë™ì˜/11_Github/mygit-1/DAT/25-1í”„ë¡œì íŠ¸/"
handle_path = os.path.join(base_path, "exolyt_result.csv")
result_csv_path = os.path.join(base_path, "exolyt_result5.csv")

# ì—‘ì…€ì—ì„œ Username ì¶”ì¶œ
handle_df = pd.read_csv(handle_path)
usernames = handle_df.iloc[:, 0].dropna().apply(lambda x: x.lstrip('@')).tolist()

print("âœ… ì‚¬ìš©ì ìˆ˜:", len(usernames))
print("ğŸ¯ ë¯¸ë¦¬ë³´ê¸°:", usernames[:])
```

```python
import json

def apply_login_cookies(driver, cookie_path):
    with open(cookie_path, 'r', encoding='utf-8') as f:
        json_data = json.load(f)
    
    cookies_raw = json_data.get('cookies', [])
    
    # ë¬¸ìì—´ì¸ ì¿ í‚¤ â†’ ê°ì²´ë¡œ ë³€í™˜
    cookies = []
    for c in cookies_raw:
        if isinstance(c, str):
            try:
                cookies.append(json.loads(c))
            except:
                continue
        else:
            cookies.append(c)

    # ì²« ë°©ë¬¸ (ì¿ í‚¤ ì„¤ì •ì„ ìœ„í•´)
    driver.get("https://www.exolyt.com")
    time.sleep(2)

    for cookie in cookies:
        try:
            cookie.pop('domain', None)  # ë„ë©”ì¸ ì¶©ëŒ ë°©ì§€
            driver.add_cookie(cookie)
        except Exception as e:
            print(f"âš ï¸ ì¿ í‚¤ ì¶”ê°€ ì‹¤íŒ¨: {cookie.get('name', 'N/A')} â†’ {e}")
```

```python
def crawl_exolyt(username):
    try:
        options = webdriver.ChromeOptions()
        options.add_argument("--no-sandbox")
        options.add_argument("--disable-dev-shm-usage")
        options.add_argument("start-maximized")
        options.add_argument("user-agent=Mozilla/5.0")

        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

        # âœ… 1. ë¡œê·¸ì¸ ì¿ í‚¤ ì ìš©
        cookie_path = os.path.join(base_path, "exolyt_login_state.json")
        driver.get("https://www.exolyt.com")
        time.sleep(2)
        apply_login_cookies(driver, cookie_path)

        # âœ… 2. ìœ ì € í˜ì´ì§€ ì´ë™
        driver.get(f"https://exolyt.com/user/tiktok/{username}")
        time.sleep(random.uniform(4, 6))

        # âœ… 3. ë¡œë”© ì™„ë£Œ ëŒ€ê¸°
        WebDriverWait(driver, 10).until(
            EC.invisibility_of_element_located((By.CLASS_NAME, "ds-fade-in"))
        )

        # âœ… 4. íŒì—… ë‹«ê¸° (ìˆìœ¼ë©´ ë‹«ê³ , ì—†ìœ¼ë©´ í†µê³¼)
        try:
            close_btn = WebDriverWait(driver, 5).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div[3]/div/svg'))
            )
            driver.execute_script("arguments[0].click();", close_btn)
            print("âœ… íŒì—… ë‹«ê¸° ì™„ë£Œ")
            time.sleep(2)
        except Exception:
            print("â„¹ï¸ íŒì—… ì—†ìŒ â†’ ê·¸ëŒ€ë¡œ ì§„í–‰")

        # âœ… ì´ ì˜ìƒ ìˆ˜ ì¶”ì¶œ
        try:
            videos_elem = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[1]/div[2]/div[2]/div[3]/div/div[1]/div/div[1]/div[1]/div[2]/div[3]/div[3]/div[2]/div[1]')
            total_videos = videos_elem.text.strip()
        except Exception as e:
            print(f"âŒ ì˜ìƒ ê°œìˆ˜ ì¶”ì¶œ ì‹¤íŒ¨: {e}")
            total_videos = None

        data = {
        "username": username,
        "total_videos": total_videos
        }   

        print("ğŸ“Š ìµœì¢… ìˆ˜ì§‘ ë°ì´í„°:", data)
        return data

    except Exception as e:
        print(f"âŒ {username} - ì‹¤íŒ¨: {e}")
        return None

    finally:
        try:
            driver.quit()
        except:
            pass
```

```python
results = []

for i, username in enumerate(usernames):
    print(f"\n[{i+1}/{len(usernames)}] â–¶ {username}")
    result = crawl_exolyt(username)
    if result:
        results.append(result)
        print(f"âœ… {username} - ì™„ë£Œ")
    else:
        print(f"âŒ {username} - ì €ì¥ë˜ì§€ ì•ŠìŒ")
```

```python
result_df = pd.DataFrame(results)
result_df.to_csv(result_csv_path, index=False, encoding='utf-8-sig')
print(f"\nğŸ“ ì €ì¥ ì™„ë£Œ: {result_csv_path}")
```
