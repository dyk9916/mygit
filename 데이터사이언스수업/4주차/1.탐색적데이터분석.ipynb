```python
df2 = df[['hotel', 'lead_time']] # hotel과 lead_time 컬럼만 추출
df2.head()

# 각 컬럼의 속성 및 결측치 확인
# df.info()

# 각 컬럼의 통계치 확인, 수치형 변수에 대해서만 확인 가능, count, mean, std, min, 25%, 50%, 75%, max
df.describe()

# 각 컬럼의 왜도 확인, 수치형 변수에 대해서만 확인 가능
numeric_columns = df.select_dtypes(include=['number']) # 수치형 변수만 추출
numeric_columns.skew()

# 각 컬럼의 첨도 확인
numeric_columns.kurtosis()

# # 특정 변수 분포 시각화
plt.rcParams['figure.dpi'] = 300
sns.distplot(df['lead_time']) # lead_time 분포 시각화

# # 호텔 구분에 따른 lead_time 분포 차이 시각화
sns.violinplot(x="hotel", y="lead_time", data=df, inner=None, color=".8")
sns.stripplot(x="hotel", y="lead_time", data=df, size=1)
```
