### 회귀성능평가지표

R-square(결정 계수)는 회귀 모델이 종속 변수를 얼마나 잘 설명하는지 나타내는 지표이다. 값은 0~1 사이이며, 1에 가까울수록 모델이 데이터를 잘 설명한다는 의미    
한계: 독립 변수가 추가될 때 마다 결정 계수는 항상 증가하게 된다, 모델이 과적합된 것일 수도 있다.    

RMSE(평균제곱근오차)    
회귀 모델은 실제값과 예측값 간의 차이를 측정하여 정확도를 평가한다.   
RMSE는 회귀 모델의 예측값과 실제값 간 평균적인 차이를 나타내는 대표적인 지표이다.   
오차를 제곱하여 평균을 구한 후 다시 루트를 씌웠다.    
이 값이 클수록 오차가 크다는 뜻이다.    
한계: RMSE는 오차의 제곱을 사용하기 때문에 이상치의 영향을 크게 받으므로 이상치에 민감하다

MAE(평균 절대 오차)   
실제값과 예측값의 차이를 절댓값으로 변환한 후 평균을 구하는 지표이다.   
오차의 절댓값을 사용하기 때문에, 예측값과 실제값의 평균적인 차이를 직관적으로 측정할 수 있다, RMSE와 달리 오차를 제곱하지 않으므로, 큰 오차보다는 전체적인 오차 평균을 중시한다.    
장점: 해석이 직관적, 이상치에 덜 민감, 예측 모델 간 비교가 쉬움   
단점: 오차의 방향을 고려하지 않음, 오차가 큰 경우 RMSE보다 차이를 강조하지 않음

MAPE(평균 절대 백분율 오차)   
예측값과 실제값의 차이를 백분율로 변환하여 평가하는 지표이다.   
  - 스케일의 영향을 받지 않기 때문에, 서로 다른 데이터셋이나 모델 간 성능 비교에 유용하다.    

장점: 스케일에 영향을 받지 않으므로 모델 비교가 용이하다, 값이 0에 가까울수록 예측 성능이 우수하다    
단점: 실제값이 0이면 계산 불가능, 실제값이 작을 경우, 오차가 과대평가될 가능성이 있음, 과소 예측과 과대 예측 시 MAPE 값이 비대칭적으로 증가   



AIC와 BIC   
AIC는 모델의 적합도가 중요한 경우, 변수 개수가 많아도 적합도가 높은 모델을 선택하고 싶은 경우, 예측 정확도를 최대화하는 것이 목표일 때 사용하는 지표이다.   
반면 BIS는 모델의 다순성이 중요하고, 변수 개수를 최소화하고, 변수가 너무 많아서 해석이 안될 때 사용하는 지표이다.
