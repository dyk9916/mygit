```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#    (2)
np.random.seed(42)
X = np.random.rand(100, 2)  # 100 , 2 

#   ( 0,  1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA  (1 )
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X_scaled)

#   
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], label=" ", alpha=0.5)
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("PCA   ")
plt.legend()
plt.show()

# PCA    
plt.scatter(X_pca, np.zeros_like(X_pca), label="PCA  ", alpha=0.5)
plt.xlabel("PC1 ( 1)")
plt.title("PCA    (1)")
plt.legend()
plt.show()
```
