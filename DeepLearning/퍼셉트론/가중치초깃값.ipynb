```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.random.randn(1000, 100)  # 입력 데이터 1000개, 표준정규분포로 랜덤한 값 생성 
node_num = 100                  # 각 은닉층의 노드(뉴런) 수
hidden_layer_size = 5           # 은닉층이 5개
activations = {}
```

```python
# 활성화 함수로 시그모이드 함수 사용
for i in range(hidden_layer_size):
    if i != 0:
        x = activations[i-1]
        
    # 가중치 초깃값 설정
    w = np.random.randn(node_num, node_num) * 1 # 표준정규분포(평균:0, 표준편차:1)로부터 생성된 난수를 배열을 생성 
    a = np.dot(x, w)
    z = sigmoid(a)
    activations[i] = z
```

```python
for i, a in activations.items():
    plt.subplot(1, len(activations), i+1)
    plt.title(str(i+1) + "-layer")
    plt.hist(a.flatten(), 30, range=(0, 1))
```
