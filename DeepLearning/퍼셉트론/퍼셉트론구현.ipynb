```python
# AND gate

def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + w2*x2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
  
# NAND gate
def NAND(x1,x2):
  w1, w2, theta = -0.5, -0.5, -0.7
  tmp = x1*w1 + w2*x2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
  
# OR gate
def OR(x1,x2):
  w1, w2, theta = 0.1, 0.1, 0
  tmp = x1*w1 + w2*x2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
```

```python
AND(0, 0) # 0
# AND(1, 0) # 0
# AND(0, 1) # 0
# AND(1, 1) # 1
```

```python
OR(0, 0) # 0
OR(1, 0) # 1
OR(0, 1) # 1
OR(1, 1) # 1
```
