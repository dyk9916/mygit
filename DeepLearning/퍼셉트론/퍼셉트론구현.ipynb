```python
import numpy as np
```

```python
# AND gate

def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + w2*x2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
  
# NAND gate
def NAND(x1,x2):
  w1, w2, theta = -0.5, -0.5, -0.7
  tmp = x1*w1 + w2*x2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
  
# OR gate
def OR(x1,x2):
  w1, w2, theta = 0.1, 0.1, 0
  tmp = x1*w1 + w2*x2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
```

```python
AND(0, 0) # 0
# AND(1, 0) # 0
# AND(0, 1) # 0
# AND(1, 1) # 1
```

```python
OR(0, 0) # 0
OR(1, 0) # 1
OR(0, 1) # 1
OR(1, 1) # 1
```

```python
# AND 게이트의 theta를 -b(편향)로 수정
def AND(x1, x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  b = -0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
AND(0,1)
```

```python
def XOR(x1,x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1,s2)
  return y
```

```python
XOR(0,0) # 0
# XOR(1,0) # 1
# XOR(0,1)  # 1
# XOR(1,1) # 0
```
