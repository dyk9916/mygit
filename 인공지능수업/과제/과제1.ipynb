### 1. 인공지능을 학습할 때 알아야 할 주요 모델과 기술 설명 

1. CNN  
CNN(Convolutional Neural Network)은 주로 이미지 처리에 사용되는 신경망 구조입니다.  
CNN은 일반적인 완전연결 신경망과는 달리, spatial data의 공간적 구조를 효율적으로 학습하기 위해 합성곱 연산(Convolution)을 사용합니다. 합성곱 층에서는 커널(필터)가 spatial data의 일부 픽셀과의 합성곱 연산을 통해 특정 패턴을 추출합니다. 이를 통해 선, 엣지, 텍스처와 같은 다양한 특징을 추출하게 됩니다. 커널을 사용해 패턴을 추출하는 이유는 spatial data는 인접한 픽셀들간의 관계가 곧 패턴을 나타내기 때문에 spatial data를 1차원 배열로 변환 후에 신경망 학습을 하게 되면 패턴 추출 하는 것이 불가능해지는 문제가 발생할 수 있습니다. 따라서 픽셀들간의 물리적인 위치를 고정해둔채 커널을 이동시켜가면서 신경망 학습을 합니다.  
CNN의 학습은 일반적인 신경망 구조와 동일하게 backpropagation 과정을 거치면서 loss를 줄이는 방향으로 weight를 갱신하는 방법입니다. 
이러한 backpropagation 과정을 여러번 거치는 경사하강법을 통해 학습 데이터에 최적화된 모델을 만들게 됩니다.  
CNN은 복잡한 패턴을 spatial data로부터 추출하는데 뛰어난 모델로써 이미지 분류, 물체 탐지, 얼굴 인식 등 다양한 분야에서 높은 성능을 자랑합니다.    


2. RNN  
RNN(Recurrent Neural Network)는 Sequential data를 처리하는데 특화된 신경망 구조입니다.  
RNN은 일반적인 완전연결 신경망과 달리, Time Series나 자연어와 같이 sequence가 존재하는 데이터를 학습하는 데 유리합니다. RNN은 은닉상태(hidden state)를 사용하여 이전 time step까지의 정보와 현재 time step의 정보를 Weighted Sum하여 현재 time step의 은닉상태를 만들어내고 이를 time step으로 전달합니다. RNN에서는 하나의 recurrent 구조가 사용되고 이 recurrent 구조에 data가 차례대로 전달되는 방식으로 추론과 학습이 진행됩니다. 이 과정에서 RNN은 매 시점에 나타날 정보를 출력하게 됩니다. 
RNN의 학습 과정은 일반적인 신경망 구조와 동일하게 backpropagation 과정을 거치지만, sequence data가 너무 길어지는 경우 앞 쪽에 위치한 data에 대한 학습은 이뤄지지 않는 문제가 발생할 수 있습니다. 이러한 문제의 원인은 기울기 소실입니다.   
신경망 구조가 학습되는 과정은 가중치에 대한 손실함수의 기울기를 활용해 가중치를 갱신합니다. 그러나 학습 도중 가중치에 대한 손실함수의 기울기가 층을 거듭할 수록 매우 작아지게 되는데 이러한 현상을 기울기 소실이라고 부릅니다. 기울기가 0에 가까워지게 되면 가중치를 갱신하는 것이 불가능해지기 때문에 학습이 제대로 이루어지지 않게 되는 것입니다. 이러한 이유로 기울기 소실은 학습을 하는데 방해가 됩니다. 따라서 RNN은 학습 시 기울기 소실 문제를 완화하기 위해 BPTT 방식을 사용합니다.  
RNN은 자연어처리, 음성 인식, 시계열 예측 등 다양한 분야에서 높은 성능을 자랑하며 LSTM, Seq2Seq, LLM의 뿌리라고 할 수 있습니다.
