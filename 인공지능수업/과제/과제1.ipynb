### 1. 인공지능을 학습할 때 알아야 할 주요 모델과 기술 설명 

1. CNN  
CNN(Convolutional Neural Network)은 주로 이미지 처리에 사용되는 신경망 구조입니다.  
CNN은 일반적인 완전연결 신경망과는 달리, spatial data의 공간적 구조를 효율적으로 학습하기 위해 합성곱 연산(Convolution)을 사용합니다. 합성곱 층에서는 커널(필터)가 spatial data의 일부 픽셀과의 합성곱 연산을 통해 특정 패턴을 추출합니다. 이를 통해 선, 엣지, 텍스처와 같은 다양한 특징을 추출하게 됩니다. 커널을 사용해 패턴을 추출하는 이유는 spatial data는 인접한 픽셀들간의 관계가 곧 패턴을 나타내기 때문에 spatial data를 1차원 배열로 변환 후에 신경망 학습을 하게 되면 패턴 추출 하는 것이 불가능해지는 문제가 발생할 수 있습니다. 따라서 픽셀들간의 물리적인 위치를 고정해둔채 커널을 이동시켜가면서 신경망 학습을 합니다.  
CNN의 학습은 일반적인 신경망 구조와 동일하게 backpropagation 과정을 거치면서 loss를 줄이는 방향으로 weight를 갱신하는 방법입니다. 
이러한 backpropagation 과정을 여러번 거치는 경사하강법을 통해 학습 데이터에 최적화된 모델을 만들게 됩니다.  
CNN은 복잡한 패턴을 spatial data로부터 추출하는데 뛰어난 모델로써 이미지 분류, 물체 탐지, 얼굴 인식 등 다양한 분야에서 높은 성능을 자랑합니다.    


2. RNN  
RNN(Recurrent Neural Network)는 Sequential data를 처리하는데 특화된 신경망 구조입니다.  
RNN은 일반적인 완전연결 신경망과 달리, Time Series나 자연어와 같이 sequence가 존재하는 데이터를 학습하는 데 유리합니다. RNN은 은닉상태(hidden state)를 사용하여 이전 time step까지의 정보와 현재 time step의 정보를 Weighted Sum하여 현재 time step의 은닉상태를 만들어내고 이를 time step으로 전달합니다. RNN에서는 하나의 recurrent 구조가 사용되고 이 recurrent 구조에 data가 차례대로 전달되는 방식으로 추론과 학습이 진행됩니다. 이 과정에서 RNN은 매 시점에 나타날 정보를 출력하게 됩니다. 
RNN의 학습 과정은 일반적인 신경망 구조와 동일하게 backpropagation 과정을 거치지만, sequence data가 너무 길어지는 경우 앞 쪽에 위치한 data에 대한 학습은 이뤄지지 않는 문제가 발생할 수 있습니다. 이러한 문제의 원인은 기울기 소실입니다.   
신경망 구조가 학습되는 과정은 가중치에 대한 손실함수의 기울기를 활용해 가중치를 갱신합니다. 그러나 학습 도중 가중치에 대한 손실함수의 기울기가 층을 거듭할 수록 매우 작아지게 되는데 이러한 현상을 기울기 소실이라고 부릅니다. 기울기가 0에 가까워지게 되면 가중치를 갱신하는 것이 불가능해지기 때문에 학습이 제대로 이루어지지 않게 되는 것입니다. 이러한 이유로 기울기 소실은 학습을 하는데 방해가 됩니다. 따라서 RNN은 학습 시 기울기 소실 문제를 완화하기 위해 BPTT 방식을 사용합니다.  
RNN은 자연어처리, 음성 인식, 시계열 예측 등 다양한 분야에서 높은 성능을 자랑하며 LSTM, Seq2Seq, LLM의 뿌리라고 할 수 있습니다.  


### 2. 머신러닝의 지도학습의 대표적인 모델과 기술들 정리  

1. Logistic Regression   
Logistic Regression는 이진 분류 문제에서 널리 사용되는 대표적인 기계 학습 알고리즘으로, 시그모이드 함수를 사용해 입력 데이터를 0과 1 사이의 확률 값으로 변환하여 특정 클래스에 속할 확률을 예측합니다. 예측 확률이 0.5 이상이면 클래스 1, 그렇지 않으면 클래스 0으로 분류합니다.  
학습 시에는 정답과 예측 확률을 이용한 Binary Cross-Entropy Loss를 사용해, 이 Loss에 대한 Weight 기울기로 Weight를 갱신합니다. Weight 갱신을 통해 정답과 예측 확률 간의 오차를 줄여나가면서 모델을 데이터에 대해 학습시킵니다. Logistic Regression은 간단하면서도 좋은 성능을 보이는 해석력 또한 높안 많이 사용됩니다. 

2. K-NN  
K-NN은 새로운 데이터 포인트가 주어졌을 때, 해당 데이터 포인트에서 가장 가까운 K개의 이웃 데이터를 찾고, 이 이웃들의 다수결 투표를 통해 클래스를 결정하거나 평균을 통해 값을 예측하는 모델입니다. K-NN의 성능은 K 값의 설정에 따라 달라지며, 너무 작은 K 값은 모델이 이상치에 민감해질 수 있고, K이 큰 경우에는 분류 성능 자체가 떨어질 수 있습니다. K-NN은 차원이 작은 데이터를 사용할 때 직관적으로 시각화하기 쉽다는 장점이 있습니다. 

3. Decision Tree  
Decision Tree는 트리 기법을 활용해 분류, 회귀 문제를 해결하는 모델로 루트 노드에서 시작해 각 노드에서 특정 조건을 기준으로 데이터를 분할하면서 최종적으로 리프 노드에 노달합니다. 각 노드의 하나의 변수를 선택해 해당 변수의 값에 따라 데이터를 나누는 방식으로 예측합니다.  결정 트리 노드 분할 기준으로는 Gini Index, Entropy 등을 사용하며, 이러한 기준을 통해 가장 효과적인 데이터 분류를 진행합니다. 

### 3. scikit-learn에서 제공하는 data를 사용하여 분류와 회귀 알고리즘 중에서 하나를 선택하여 실습하고 프로그램을 제출 

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
# 데이터 생성
X, y = make_blobs(n_samples=1000, centers=6, random_state=42) # n_samples: 데이터 개수, centers: 군집 개수

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# KNN 모델 학습
knn = KNeighborsClassifier(n_neighbors=5) # n_neighbors: 이웃 개수
knn.fit(X_train, y_train)

# 예측
y_pred = knn.predict(X_test)

# 결과 출력
print(classification_report(y_test, y_pred))

```

```python
# 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='viridis') # 실제 정답 데이터, c=y, 동그라미로 표시

plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred, s=50, cmap='viridis', marker='x') # 예측 데이터, c=y_pred, 엑스로 표시
plt.title('KNN Classification Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
